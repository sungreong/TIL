## tabular cat + num

```python
import tensorflow as tf
exported_path =\
'/tmp/testmodels/uniform_average_numcat_classifier_ensemble_baseline/1577623078/'
model = tf.saved_model.load(export_dir=exported_path)
```

```python
objcol = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
numcol = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
```

```python
input_col = ['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays', 'previous', 'poutcome']
```

```python
list_value = [35.0, 'technician', 'divorced', 'secondary', 'no', 473.0, 'yes', 'yes', 'unknown', 26.0, 'may', 434.0, 4.0, -1.0, 0.0, 'unknown']
```

```python
def _bytes_feature(value):
    """string / byte   byte list ."""
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))

def _float_feature(value):
    """float / double   float list ."""
    return tf.train.Feature(float_list=tf.train.FloatList(value=[value]))

def _int64_feature(value):
    """bool / enum / int / uint   int64 list ."""
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))
```

```python
def predict(model , x  ):
    example = tf.train.Example()
    #feature = {}
    for idx , value in enumerate(x) :
        if input_col[idx] in  objcol :
            #feature[input_col[idx]] = _bytes_feature(value.encode('utf-8'))
            example.features.feature[input_col[idx]].bytes_list.value.extend( [value.encode("utf-8") ])
        elif input_col[idx] in  numcol :
            #feature[input_col[idx]] = _float_feature(value)
            example.features.feature[input_col[idx]].float_list.value.extend([value])
    return model.signatures["predict"](examples=tf.constant([example.SerializeToString()]))
```

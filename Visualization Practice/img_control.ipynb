```python
from sklearn.preprocessing import MinMaxScaler

data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
scaler = MinMaxScaler(feature_range=(-1,1))
print(scaler.fit(data))
```

```python
print(scaler.transform(data))
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
print(scaler.fit(data))
print(scaler.transform(data))
```

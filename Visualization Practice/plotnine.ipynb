```python
(ggplot(pokemon) + 
 aes("Attack", "Defense",color="Type 1") + 
 geom_point())
```

```python
import matplotlib.font_manager as fm
path = 'C:/Anaconda3/envs/etc/Lib/site-packages/pytagcloud/fonts/TmonMonsori.ttf'
fontprop = fm.FontProperties(fname=path, size=10)

(ggplot(pokemon)
 + aes('Attack','Defense', color='Legendary')
 + geom_point()
 + ggtitle('Legendary   HP    ')
 + theme(text=element_text(fontproperties=fontprop))
)
```

```python
(ggplot(pokemon) 
+aes("Attack","Defense", color="HP")
+geom_point()
+ggtitle("HP    ")
+theme(text=element_text(fontproperties=fontprop))
)
```

```python
p= ggplot(aes(x="Attack", y = "Defense"), pokemon)
```

```python
p+geom_point()
```

```python
p+geom_point(aes(color="Type 1"))+stat_smooth(method="lm") + facet_wrap("Legendary")
```

```python
p+geom_point(aes(color="Speed"))+ scale_color_gradient(low="blue", high="red")
```

```python
p+geom_point(aes(size="Legendary", color="HP"))
```

```python
(ggplot(pokemon)
 + geom_col(aes("Type 1", "Attack" , fill="Legendary", na_rm=True ))
)
```

```python
(ggplot(pokemon)
+aes(x="Type 1" , y="HP", fill="Legendary")
+geom_col()
+theme(text=element_text() ,axis_text_x=element_text(rotation=60)
))
```

```python
(ggplot(pokemon)
+aes(x="Type 1" , y= "HP")
+geom_boxplot(outlier_color="red", outlier_shape="s" ,na_rm=True)
+theme(text=element_text() ,axis_text_x=element_text(rotation=60))
)
```

```python
(ggplot(pokemon, aes(x="Type 1", y="Attack", fill="Legendary"))
+ geom_bar(stat="identity", position="dodge")+
theme(text=element_text(), axis_text_x=element_text(rotation=60)))
```

```python
(ggplot(pokemon, aes(x="Generation", y="Attack", fill="Legendary"))
+ geom_bar(stat="identity", position="dodge")+
theme(text=element_text(), axis_text_x=element_text(rotation=60)))
```

```python
pokemon = pokemon.rename(index= str, columns={"Type 1": "type1", "Type 2":'type2'})
pokemon.head()
```

```python
(ggplot(pokemon)
+aes(x="type1", y="Total")
+geom_col(fill="darkslateblue")
+coord_flip()
+theme(figure_size=(8,6)))
```

```python
(ggplot(pokemon)
+aes(x="Generation", y="Total",fill="Legendary")
+geom_col()
+coord_flip()
+theme(figure_size=(8,6)))
```

```python
(ggplot(pokemon)
+aes("Attack","Defense")
+aes(color="Legendary")
+geom_point()
+facet_wrap("type1"))
```

```python
(ggplot(pokemon)
+aes("Attack","Defense")
+aes(color="factor(Legendary)")
+geom_point()
+facet_wrap("Generation"))
```

```python
(ggplot(pokemon)
+aes("Attack","Defense")
+geom_bin2d(bins=10))
```

```python
(ggplot(pokemon)
+aes("HP","Speed")
+geom_bin2d(bins=10)
)
```

```python
(ggplot(pokemon)
 + aes('type1', 'Defense')
 + geom_boxplot(outlier_color="red") 
 + theme(text=element_text(),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon)
+aes("type1", "Attack")
+aes(color="Legendary")
+geom_boxplot()
+theme(text=element_text(),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon)
+aes("type1","type2")
+geom_bin2d(bins=3)
 +theme(text=element_text(),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon, aes(x='type1', y='HP',fill="factor(type2)"))
 + geom_col()
 + theme(text=element_text(),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon)
 + aes('type1', 'HP')
 + geom_violin() 
 + theme(text=element_text(),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon)
 + aes('type1', 'Attack', fill='Legendary')
 + geom_violin() 
 + theme(text=element_text(),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon)
 + aes('type1', 'Speed', fill='Legendary')
 + geom_jitter() 
 + theme(text=element_text(),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon)
 + aes('Attack', 'Defense', fill='Legendary')
 + geom_jitter() 
 + geom_rug() 
 + theme(text=element_text(),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon)
 + aes('Attack', 'Defense', fill='Legendary')
 + geom_area(position="stack")  
 + theme(text=element_text(),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon)
 + aes('Attack')
 + geom_density()  
 + theme(text=element_text(),
        axis_text_x=element_text(rotation=70))
 +facet_wrap("type1")
)
```

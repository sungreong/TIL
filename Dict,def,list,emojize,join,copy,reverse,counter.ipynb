## min , max index ì°¾ê¸° 

```
lst = [40 ,10 ,12, 13]
def minindex(list) :
    return min( range( len(list )) , key = list.__getitem__ ) 
def maxindex(list) :
    return max( range( len(list )) , key = list.__getitem__ ) 


print(minindex( lst ))
print(maxindex( lst ))
```

## copy deepcopy êµ¬ë¶„ì§€ì–´ì„œ ì‚¬ìš©í•˜ê¸°

```
import copy


"""
shallow copy
"""
a = [1, [1, 2, 3]]
b = copy.copy(a)    # shallow copy ë°œìƒ
print(b)    # [1, [1, 2, 3]] ì¶œë ¥
b[0] = 100
print(b)    # [100, [1, 2, 3]] ì¶œë ¥,
print(a)    # [1, [1, 2, 3]] ì¶œë ¥, shallow copy ê°€ ë°œìƒí•´ ë³µì‚¬ëœ ë¦¬ìŠ¤íŠ¸ëŠ” ë³„ë„ì˜ ê°ì²´ì´ë¯€ë¡œ itemì„ ìˆ˜ì •í•˜ë©´ ë³µì‚¬ë³¸ë§Œ ìˆ˜ì •ëœë‹¤. (immutable ê°ì²´ì˜ ê²½ìš°)

c = copy.copy(a)
c[1].append(4)    # ë¦¬ìŠ¤íŠ¸ì˜ ë‘ë²ˆì§¸ item(ë‚´ë¶€ë¦¬ìŠ¤íŠ¸)ì— 4ë¥¼ ì¶”ê°€
print(c)    # [1, [1, 2, 3, 4]] ì¶œë ¥
print(a)    # [1, [1, 2, 3, 4]] ì¶œë ¥, aê°€ cì™€ ë˜‘ê°™ì´ ìˆ˜ì •ëœ ì´ìœ ëŠ” ë¦¬ìŠ¤íŠ¸ì˜ item ë‚´ë¶€ì˜ ê°ì²´ëŠ” ë™ì¼í•œ ê°ì²´ì´ë¯€ë¡œ mutableí•œ ë¦¬ìŠ¤íŠ¸ë¥¼ ìˆ˜ì •í• ë•ŒëŠ” ë‘˜ë‹¤ ê°’ì´ ë³€ê²½ë¨

"""
deepcopy
"""

a = [1, [1, 2, 3]]
b = copy.deepcopy(a)    # deep copy ì‹¤í–‰
print(b)    # [1, [1, 2, 3]] ì¶œë ¥
b[0] = 100
b[1].append(4)
print(b)    # [100, [1, 2, 3, 4]] ì¶œë ¥
print(a)    # [1, [1, 2, 3]] ì¶œë ¥
```
